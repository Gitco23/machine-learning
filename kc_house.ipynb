```python
sqft_basements = df[df['sqft_basement'] == 0]
sqft_basements
```

```python

sqft_basement_mean=df[df['sqft_basement'] != 0]['sqft_basement'].mean()
sqft_basement_mean
```

```python
df.replace({'sqft_basement': 0}, sqft_basement_mean, inplace=True)
df.head()
```

```python
yr_renovated_mean = df[df['yr_renovated'] != 0]['yr_renovated'].mean()
df.replace({'yr_renovated': 0}, yr_renovated_mean, inplace=True)
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df = scaler.fit_transform(df)
df
```

```python
import xgboost as xgb
import numpy as np
```

```python
X = data.iloc[:, 3:].values
y = data['price'].values
```

```python
np.shape(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
xg_reg = xgb.XGBRegressor(colsample_bytree=0.8, learning_rate=0.1, max_depth=6, n_estimators=1000, verbosity=3)
xg_reg.fit(X_train, y_train)
```

```python
y_pred = xg_reg.predict(X_test)
```

```python
from sklearn import metrics
```

```python
metrics.mean_absolute_error(y_test, y_pred)
```

```python
metrics.r2_score(y_test, y_pred)
```

```python
xgb.plot_importance(xg_reg)
```

```python
from sklearn.svm import SVR
s = SVR()
s.fit(X_train, y_train)
y_pred = s.predict(X_test)
```

```python
metrics.mean_absolute_error(y_test, y_pred)
```
